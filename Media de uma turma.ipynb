```python
#Giovanni Andretta Carbonero

#Listas
nome = []
nota = []
aprovados = []
reprovados = []

#Função para calcular a media da turma
def calcula():
    media = sum(nota)
    media = media / len(nome)
    
    return media

#Função que coleta as infos principais
def infos():
    
    opt = "S"
    
    while opt == "S":
        aluno = str(input("Digite o nome do aluno: "))
        nome.append(aluno)

        aux = float(input("Digite a nota do aluno: "))
        nota.append(aux)

        opt = str(input("Deseja digitar a nota de outro aluno? ('S' para sim): "))
        opt = opt.upper()
        opt = opt[0]
        
#Função para exibir todos os resultados
def exibi():
        print("\n\nInformações: \n")
        
        print("="*50)
        
        print("\nAlunos cadastrados e suas notas: \n")
        
        for nomes, notas in zip(nome, nota):
            print("{}: {}".format(nomes, notas))
            
        print("="*30)
            
        media = calcula()
        
        print("\nMédia da turma: %.2f\n" % (media))
        
        print("="*30)
        
        print("Aprovados:\n")
        for i in aprovados:
            print(i)
        
        print("\nReprovados:\n")
        for j in reprovados:
            print(j)
            
        print("="*50)

#Função para verificar os alunos aprovados e reprovados
def verifica():
    for nomes, notas in zip(nome, nota):
        if notas >= 6:
            aprovados.append(nomes)
            
        else:
            reprovados.append(nomes)
                 
#Função principal       
def main():
    infos()
    calcula()
    verifica()
    exibi()
    
main()
```
